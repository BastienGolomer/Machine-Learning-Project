```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import random
```

```python
train=[]
with open('train.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        train.append(np.asarray(row)) #read rows of train.csv and change type to array
train=np.asarray(train)#creates a matrix rather than a list of array
```

```python
#creates input and outputs in a useable shape
y=train[1:,1]
obs=train[0,:]
train=train[1:,2:]
```

```python
#this block is not necessary but can be useful if we decide to not look at the -999 values

missing=[]#number of the columns with missing values
for i in range (0,len(train[1,:])-1):
    for j in range (0,len(train[:,1])-1):
        if train[j,i] =='-999':
            if not (i in missing):
                missing.append(i)
            train[j,i]=None#replace values with none
```

```python
print(len(train[1,:]),len(train[:,1]))
```

```python
#drop missing collumns
missing=np.sort(missing)[::-1]
for i in missing:
    train=np.delete(train,i,1)
```
